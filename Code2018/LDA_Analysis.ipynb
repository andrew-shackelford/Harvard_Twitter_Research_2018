### LDA Analysis on Words in Tweets (not just hashtags)
We first run an LDA analysis (using nltk and gensim libraries) on tweets from October 18th, to find the top 5 word clusters.

```python
# import nltk and other libraries for LDA analysis; define clean function
import nltk
import string
from nltk.corpus import stopwords 

from nltk.stem.wordnet import WordNetLemmatizer
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    no_rts = punc_free.replace('rt ', '')
    normalized = " ".join(lemma.lemmatize(word) for word in no_rts.split())
    return normalized
```

```python
# clean tweets
tweets_clean = [clean(tweet).split() for tweet in tweet_list]   
```

```python
import gensim
from gensim import corpora

# Creating the term dictionary of our courpus, where every unique term is assigned an index.
dictionary = corpora.Dictionary(tweets_clean)

# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
tweet_term_matrix = [dictionary.doc2bow(tweet) for tweet in tweets_clean]
```

```python
# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel

# Running and Trainign LDA model on the document term matrix.
ldamodel = Lda(tweet_term_matrix, num_topics=5, id2word = dictionary, passes=10)
```

```python
print(ldamodel.print_topics(num_topics=5, num_words=5))
```

This doesn't seem to be very helpful for gauging sentiment (feel free to tweak the number of words showed for each cluster to see more results). There don't seem to be any strongly-partisan words that come up (except "maga"). We should probably also ignore any instance of "brexit". Sentiment analysis may be better done on hashtags...

```python
for i in range(5):
    i = i+1
    print("%i Clusters:" %i, ldamodel.print_topics(num_topics=i))
    print()
```

### LDA Analysis on Hashtags

Next I keep only the hashtags in the tweets, and run another LDA analysis on that.

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()

# this clean function keeps only hashtags
def clean_to_hashtags(doc):
    stop_free_hashtags = " ".join([i for i in doc.lower().split() if (i not in stop) and (i in hashtags)])
    punc_free = ''.join(ch for ch in stop_free_hashtags if ch not in exclude)
    no_rts = punc_free.replace('rt ', '')
    normalized = " ".join(lemma.lemmatize(word) for word in no_rts.split())
    return normalized
```

```python
#clean tweets
tweets_clean_h = [clean_to_hashtags(tweet).split() for tweet in tweet_list]   
```

```python
import gensim
from gensim import corpora

# Creating the term dictionary of our courpus, where every unique term is assigned an index.
dictionary_h = corpora.Dictionary(tweets_clean_h)

# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
tweet_term_matrix_h = [dictionary_h.doc2bow(tweet) for tweet in tweets_clean_h]
```

```python
# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel

# Running and Trainign LDA model on the document term matrix.
ldamodel_h = Lda(tweet_term_matrix_h, num_topics=5, id2word = dictionary_h, passes=10)
```

```python
for i in range(5):
    i = i+1
    print("%i Clusters:" %i, ldamodel_h.print_topics(num_topics=i))
    print()
```

Interestingly enough, the LDA clustering finds it hard to separate clearly-partisan tweets, although it is interesting to see which hashtags are most significant (the weightage is seen by the number beside the hashtag) to each topic. Generally, the more right-wing keywords are weighted more heavily in the topics ("maga" comes up a lot; "voteredtosaveamerica" "walkaway" also extremely common right-wing tweets)

### Perplexity, Coherence and LDA Visualization

```python
# Compute Perplexity: a measure of how good the model is. lower the better.
print('\nPerplexity: ', ldamodel.log_perplexity(tweet_term_matrix)) 

# Compute Coherence Score
coherence_model_lda1 = gensim.models.CoherenceModel(model=ldamodel, texts=tweets_clean, dictionary=dictionary, coherence='c_v')
coherence_lda1 = coherence_model_lda1.get_coherence()
print('\nCoherence Score: ', coherence_lda1)
```

```python
# Compute Perplexity: a measure of how good the model is. lower the better.
print('\nPerplexity: ', ldamodel_h.log_perplexity(tweet_term_matrix_h)) 

# Compute Coherence Score
coherence_model_lda = gensim.models.CoherenceModel(model=ldamodel_h, texts=tweets_clean_h, dictionary=dictionary_h, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

Looks like the perplexity for the tweets without the hashtags is better, but the c_v coherence for the tweets with the hashtags is better. 
Note that I've only trained the model with 10 iterations through each batch, and the dataset is only a randomized batch for one particular day, Oct 18th. This took a long time, but to get better results it would probably be best to do at least 50 iterations one ach.
